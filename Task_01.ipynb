```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/API_SP.POP.TOTL_DS2_en_csv_v2_34.csv',skiprows=4)
data.head()
```

```python
india = data[data['Country Name']=='India']
population= india.loc[:,'2010':'2020'].T
population.columns=['Population']
population
```

```python
plt.figure()
plt.bar(population.index,population['Population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India (2010-2020)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.hist(population['Population'])
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution of India')
plt.show()
```
